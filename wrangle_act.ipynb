### Issue #2,3: Dog stages with no data represented as 'None' while other columns with no data represented as 'Nan'. The dog stages in different columns while there could be one column with each row having its own dog stage

#### Define
- Replace None with np.nan
- Merge the four columns representing dog stages into one column

#### Code

```python
#Replacing NOne in the dataframe with np.nan
Dog_stages = ['doggo', 'floofer', 'pupper', 'puppo']
df_clean[Dog_stages] = df_clean[Dog_stages].replace('None', np.nan)
```

```python
def join_column(x):
    return ','.join(x.dropna().astype(str))
```

```python
df_clean['Dog_stages'] = df_clean[Dog_stages].apply(join_column, axis=1)
```

```python
df_clean['Dog_stages'] = df_clean['Dog_stages'].replace('', np.nan)
```

```python
#Drop these columns after creating one column that represents them all
df_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1, inplace=True)
```

### Conclusion#1
- Based on the scatter plot, it shows there's a negative relationship, almost no relationship at all between Retweets count and the dogs' rating. The ratings the dogs were given did not really affect the number of times a tweet/post was retweeted. This signifies that people who retweeted the posts did so based on their own judgement of how the dogs appealed to their eyes.

### Conclusion#2
- The increase in the number of retweets increased over time although the change was not as constant as that of favorite count. This could maybe signify that the WeRateDogs account got more followers most of who reacted to the tweets by liking and not retweeting

### Conclusion#3
- The stacked histogram shows that the number of favorite counts is higher than that of reweet count meaning that most of the tweets received likes from people and not retweets even though there's also a high number of retweets
